### Clean up the data

###### Drop rows of NaN

Required for changing the data type of 'Month' column to int data type.

```python
# To display rows with one or more NaN values in pandas dataframe

# Using isna() and any() with axis parameter

nan_df = all_data[all_data.isna().any(axis=1)]

nan_df.head()
```

```python
# we can use isnull() function as well with any(axis=1)

all_data[all_data.isnull().any(axis = 1)].head()
```

```python
# Dropping rows when all the values are NA in a row.

all_data = all_data.dropna(how = 'all')
```

**how{‘any’, ‘all’}**, default ‘any’
Determine if row or column is removed from DataFrame, when we have at least one NA or all NA.

**‘any’** : If any NA values are present, drop that row or column.

**‘all’** : If all values are NA, drop that row or column.

```python
# Checking it is deleted or not (but it will show for any NaN value)

all_data.isnull().sum() # Now go and try to convert the Month column data type
```

##### Find 'or' in 'Order Date' column and delete it

Rectifying below error when trying to convert 'Month' column to int.

ValueError: invalid literal for int() with base 10: 'Or'

```python
# all_data[condition] -> condition is that the first two character of Order Date is 'or'

temp_df = all_data[all_data['Order Date'].str[0:2] == 'Or']

temp_df.head()
```

```python
# removing rows with Order Date has value in Or/dd/yyyy format

all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']

# Now, go and try to convert Month column into integer
```

##### Convert 'Quantity Ordered' and 'Price Each' columns to the correct type

```python
# Using pd.to_numeric() instead of astype(), as we have to give data type in astype()

all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])

all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

### Augment data with additional columns

#### Task 2: Add Month Column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]

all_data.head()
```

- But 'Month' column is in 'string' format. Let's convert it into 'int'

```python
# Converting Month to int

all_data['Month'] = all_data['Month'].astype('int32') 

# We've removed below two errors before converting

# ValueError: cannot convert float NaN to integer          -> I removed all the Nan values

# ValueError: invalid literal for int() with base 10: 'Or' -> removed the rows where Order Date starts with 'Or'

all_data.head()
```

- Now that we've got the Month column, Let get Sales Column as well

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each'] # TypeError: can't multiply sequence by non-int of 
                                                                          # type 'str'
all_data.head()                                                           # we've changed their data type to integer 
```

###### Another ways to add city column. Using apply() and lambda

#### Task 4: Add a State Column

```python
# one issue, few state have same city name like Portand in the West Coast and East Coast. So, it is better to grab the state
# as well

def get_state(address):
    return address.split(',')[2][1:3]

all_data['State'] = all_data['Purchase Address'].apply(lambda x: get_state(x))  # x is the cell content

all_data.head()
```

- **State** column is added.

```python
# But we don't want separate City and State column, So dropping them

all_data.drop(['City','State'], axis = 1, inplace = True)

all_data.head()
```

#### we wanted City column in below format

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    #return address.split(',')[2][1:3]                 
    return address.split(',')[2].split(' ')[1]

# all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' (' + get_state(x) + ')')
all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")

all_data.head()
```

- **City** column is added with attached state info.

### Question 4: What products are most often sold together?

```python
all_data.head()
```

- **Logic**: We can look for Products with same order id. like (176560) That means, they were ordered together. or better to find out duplicate 'Order ID'.

##### Notes for df.duplicated()

DataFrame.duplicated(subset=None, keep='first')

Return boolean Series denoting duplicate rows.

Parameters
subset   column label or sequence of labels, optional
         Only consider certain columns for identifying duplicates, by default use all of the columns.

keep    {‘first’, ‘last’, False}, default ‘first’
        Determines which duplicates (if any) to mark.

   - first : Mark duplicates as True except for the first occurrence.

   - last : Mark duplicates as True except for the last occurrence.

   - False : Mark all duplicates as True.

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df.head(20)
```

```python
# Let's add a column with it's values of Products with same order id join together

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df.head()
```

**Let's drop-out duplicate occurrence of same-pair.**

```python
df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head(10)
```

- ##### These are the Products sold together.

### Question 4 extension: Count the pair, which occcurs most together?

* Search from StackOverFlow*

```python
from itertools import combinations
from collections import Counter

# create an object of Counter class
count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))  # this will search for most occurrence of pair of two
    
print(count)
```

- iPhone and Lightning Chargind Cable, pair is the most occurence.

```python
# For readable

count.most_common(10)
```

```python
# For more readable

for key,value in count.most_common(10):
    print(f"{key}, : {value}")
```

```python
# For most occurence of pair of three

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,3)))  # this will search for most occurrence of pair of two
    
for key,value in count.most_common(10):
    print(f"{key}, : {value}")
```
